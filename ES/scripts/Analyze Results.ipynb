```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

# Back Tracking
## 3x3

```python
resultFile="../ESResults/BT/SF/3x3/"
x = list()
y = list()
e = list()
for color in range(2,9):
    x.append(color)
    df = pd.read_csv(resultFile+str(color)+'.csv',header=None)
    y.append(df.mean(axis=0)[1])
    e.append(df.std(axis=0, ddof=0)[1])
plt.xticks(x)
plt.errorbar(x, y, e, marker='o')
plt.xlabel("Number of Colors")
plt.ylabel("time in seconds")
plt.title("Back Tracking 3x3")
plt.show()
```
