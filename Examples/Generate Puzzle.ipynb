## Generating Square EMP
Generates random puzzle and returns the puzzle and id string representing the puzzle generated 

                id example: "EMPType-colors-gridSizeX-gridSizeY-seed-rotate"

grid: can be a tuple or a list. tuple (m,n) will generate a puzzle of m x n. A list grid is used to pass a custom grid shape.

colors: number of colors used in the puzzle.

EMPType: type of the puzzle.

    'SF' square framed EMP.
    'S'  square none framed EMP.

return_: type of the return. default: 'pieces'

    'pieces' list of pieces
    'grid'   grid of puzzle


shuffle: shuffles the pieces of the puzzle randomly if True. default: True

rotate: rotates each piece randomly if True. default: True

seed: random seed used. defualt: None. random seed will be used.   

## 30 x 3 square framed EMP using tuple

```python
puzzle, id = generatePuzzle((30,3), 9, EMPType='SF', return_='grid', rotate=True, shuffle=True, seed=123)
```

## Custom grid passed to generatePuzzle()

```python
grid5=[[[ None for x in range(4)]for y in range(6)] for p in range(5)]
for x in range(10):
    grid5[random.randint(0,4)][random.randint(0,5)] = [-1,-1,-1,-1]

puzzle, id = generatePuzzle(grid5, 9, return_='grid', EMPType='SF', rotate=False, shuffle=False, seed=123)
drawPuzzle(puzzle)
```

## E EMP

```python
E_grid = [[[ None for x in range(4)]for y in range(3)] for p in range(5)]
E_grid[1][1] = E_grid[1][2] = [-1,-1,-1,-1]
E_grid[3][1] = E_grid[3][2] = [-1,-1,-1,-1]

puzzle, id = generatePuzzle(E_grid, 9, return_='grid', EMPType='SF', shuffle=False, rotate=False, seed=123)
drawPuzzle(puzzle)
```

```python
puzzle, id = generatePuzzle(E_grid, 9, return_='grid', EMPType='SF', shuffle=True, rotate=True, seed=123)
drawPuzzle(puzzle)
```
