```python
from EMP.Algorithms.MCTS_random import MCTS as mcts
from EMP.Algorithms.MCTS_discrete import MCTS as mcts2
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
number_of_parms = 10
number_of_runs    = 1000
x1 = np.zeros(number_of_parms)
x2 = np.zeros(number_of_parms)
y = np.arange(number_of_parms)
randseed = random.randint(0,100000)
size =(3,3)
    
for i in range(number_of_parms):

    for j in range(number_of_runs):
        
        parm = i    
        #print("mcts_discrete")
        pieces, id = generatePuzzle(size, 3, EMPType='S', return_='pieces', rotate=True, seed=randseed)
        start = time.time()
        mcts_d = mcts2(pieces,size, EMPType='S', rotate=True, parm=parm)
        mcts_d.mcts_search()
        end = time.time()
        #print(end - start)
        
        x1[i] += (end - start)
        
        
        parm = i
        #print("mcts_random")
        pieces, id = generatePuzzle(size, 3, EMPType='S', return_='pieces', rotate=True, seed=randseed)
        start = time.time()
        mcts_r = mcts(pieces,size, EMPType='S', rotate=True, parm=parm)
        mcts_r.mcts_search()
        end = time.time()
        #print(end - start)
        #print(end-start)
        
        x2[i] += (end - start) 
        
        
```

```python
drawPuzzle(mcts_d.grid)
```

```python
drawPuzzle(mcts_r.grid)
```
